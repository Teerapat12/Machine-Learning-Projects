It seems that people in Western Europe seems to be more happy overall than Central and Eastern Europe. Let's explore more on this topics on why people in Western Europe are happier.

```python
w_europe = dataset[dataset.Region=='Western Europe']
ec_europe = dataset[dataset.Region=='Central and Eastern Europe']
europe = pd.concat([w_europe,ec_europe],axis=0)
europe.head()
```

<h2>It's all about money</h2><br/>
<p>First, let's compare their economy and Happiness of the two region</p>

```python
sns.lmplot(data=europe,x='Economy',y='Happiness Score',hue="Region")
```

This plot show us again how economy affect happiness score. Now let's compare other aspect of the country

```python
selectCols=  ['Happiness Score','Economy','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Region']
sns.pairplot(europe[selectCols], hue='Region',size=2.5)
```

<ul><h2>From pairplot, we knows that</h2></ul>
<li>Western overall has higher economy, health, and freedom.</li>
<li>While wester people have some mix opinion about government, people in eastern and central tends have some issues with the government due to the fact that many people has low trust</li>

<h2>Let's compare these two region more in details</h2>

```python
f, axes = plt.subplots(3, 2, figsize=(16, 16))
axes = axes.flatten()
compareCols = ['Happiness Score','Economy','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)']
for i in range(len(compareCols)):
    col = compareCols[i]
    axi = axes[i]
    sns.distplot(w_europe[col],color='blue' , label='West', ax=axi)
    sns.distplot(ec_europe[col],color='green', label='Central/East',ax=axi)
    axi.legend()
```

<h3>Conclusion</h3><br>
Overall, rom this we can see that West people have higher score in every aspects. Note that all except trust have a high peak (left skew). So we can see now the reasons of why Western people are more happy. Because they have better life standard score compare to central/ europe. 

Now let's compare other region and see how they related. We will create a function so that it is easier reusing the code above

```python
def plot_compare(dataset,regions,compareCols):
    n = len(compareCols)
    f, axes = plt.subplots(ceil(n/2), 2, figsize=(16, 4*ceil(n/2)))
    axes = axes.flatten()
    #compareCols = ['Happiness Score','Economy','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)']
    for i in range(len(compareCols)):
        col = compareCols[i]
        axi = axes[i]
        for region in regions:
            this_region = dataset[dataset['Region']==region]
            sns.distplot(this_region[col], label=region, ax=axi)
        axi.legend()
```

```python
dataset['Region'].unique()
```

```python
plot_compare(dataset,['Western Europe', 'North America', 'Australia and New Zealand',
       'Middle East and Northern Africa', 'Latin America and Caribbean',
       'Southeastern Asia', 'Central and Eastern Europe', 'Eastern Asia',
       'Sub-Saharan Africa', 'Southern Asia'],['Happiness Score','Economy','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)'])
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.isnull().sum()
```

```python
train.isnull().count()
```

<h3>Age has 177 missing value and Cabin has up to 687 missing value!</h3><br/>
We may remove Cabin, but I want to explore whether people with Cabin has more chance of survival or not
For Age, I will fill in median for the missing value.

```python
train['Age'] = train['Age'].fillna(train['Age'].median())
test['Age'] = test['Age'].fillna(test['Age'].median())
#Define new features
train['Fam'] = train['SibSp']+train.Parch
test['Fam'] = test['SibSp']+test.Parch
```

```python
def hasCab(cab):
    if pd.isnull(cab):
        return 0
    else:
        return 1
    
train['HasCabin'] = train['Cabin'].apply(hasCab)
test['HasCabin'] = test['Cabin'].apply(hasCab)
train.HasCabin.value_counts()
```

```python
sns.countplot(x="HasCabin", hue="Survived", data=train)
```

<h3>People with Cabin survived more</h3><br/>
Strangely, people with Cabin has somewhat higher change to survived than one without. 
One possible idea is that people with Cabin are more rich, so they survived more
Let's explore more on whether Cabin relate with pClass and Fare


```python
sns.countplot(x="HasCabin", hue="Pclass", data=train)
```

<h3>Next question is that if having Cabin means more chance comparing to one in the same Pclass</h3><br/>
HasCabin means that they are likely to be in high Pclass, so what given Pclass, can we kind of guess survival based on HasCabin.

```python
low = train[train.Pclass==3]
mid = train[train.Pclass==2]
high = train[train.Pclass==1]
f, axes = plt.subplots(1, 3, figsize=(16, 4))
sns.countplot(x="HasCabin", hue="Survived", data=low, ax = axes[0])
sns.countplot(x="HasCabin", hue="Survived", data=mid, ax = axes[1])
sns.countplot(x="HasCabin", hue="Survived", data=high, ax = axes[2])
axes[0].set_title("Low Class (Pclass ==3)")
axes[1].set_title("Middle Class (Pclass ==2)")
axes[2].set_title("High Class (Pclass ==1)")
```

```python
low_cabin_survived_count = low[(low.HasCabin==1) & (low.Survived)].shape[0]
lowelse_count = low.shape[0] - low_cabin_survived_count
print("Low Class with Cabin and Survived : " +str(low_cabin_survived_count))
print("Low else : "+str(lowelse_count))

mid_cabin_survived_count = mid[(mid.HasCabin==1) & (mid.Survived)].shape[0]
midelse_count = mid.shape[0] - mid_cabin_survived_count
print("Middle Class with Cabin and Survived : " +str(mid_cabin_survived_count))
print("Middle else : "+str(midelse_count))

high_cabin_survived_count = high[(high.HasCabin==1) & (high.Survived)].shape[0]
highelse_count = high.shape[0] - high_cabin_survived_count
print("High Class with Cabin and Survived : " +str(high_cabin_survived_count))
print("High else : "+str(highelse_count))
```

<h3>Let's break this down for each classes</h3><br/>
For high class, having cabin means that you have more chance to survive.
For middle class, people with no cabin has lower change to survived.
For low class, people with no cabin also has lower chance to survived.

<h3>Last but not least, the names</h3>
Let's look at name length and word in names and see whether they are related to survival rate or not

```python
#Define our function
def name_length(name):
    return len(name)
def name_word(name):
    return len(name.split(" "))
```

```python
#Apply to dataset
train['NameLength'] = train['Name'].apply(name_length)
train['NameWord'] = train['Name'].apply(name_word)

test['NameLength'] = test['Name'].apply(name_length)
test['NameWord'] = test['Name'].apply(name_word)
```

```python
f, axes = plt.subplots(1, 2, figsize=(16, 4))

sns.distplot(train['NameLength'][train.Survived==1],color='blue' , label='Survived', ax=axes[0])
sns.distplot(train['NameLength'][train.Survived==0],color='green', label='Not Survived',ax=axes[0])

sns.distplot(train['NameWord'][train.Survived==1],color='blue' , label='Survived', ax=axes[1])
sns.distplot(train['NameWord'][train.Survived==0],color='green', label='Not Survived',ax=axes[1])
```

<h3>Name conclusion</h3><br/>
For the length longer Name word means higher survivability. <br/>
For the word, they aren't that much difference in survival rate so we are going to ignore it.

<h2>Let's build a Machine Learning Model!</h2>
Previously we some kind clean the data already, let's do some more feature scaling and clean the test set.
For the column we will choose in the model, it will be : 
<ul>
<li>Pclass</li>
<li>Sex</li>
<li>Age</li>
<li>Fare</li>
<li>HasCabin</li>
<li>NameLength</li>
</ul>


```python
train.head()
train.columns
selectCol = ['Pclass', 'Sex', 'Age', 'Fam', 'HasCabin', 'NameLength']
target = ['Survived']
X_train = train[selectCol]
y_train = train[target].values.ravel()
X_test = test[selectCol]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train['Sex'] = le.fit_transform(X_train['Sex'])
X_test['Sex'] = le.transform(X_test['Sex'])
#X_test['Fare'] = X_test['Fare'].fillna(X_test['Fare'].median())
```

```python
# Feature Scaling
apply_feature_col = []
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting Random Forest Classifier to the Training set
'''
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100,criterion='entropy',random_state=0)
classifier.fit(X_train,y_train)
'''
```

```python
# Fitting Kernel SVM to the Training set
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', C=3.8,gamma=0.0189,random_state = 0,probability=True)
classifier.fit(X_train, y_train)
```

```python
# Fitting Naive Bayes to the Training set
'''
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
'''
```

```python
# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 20,scoring='accuracy')
accuracies.mean()
```

```python
y_pred = classifier.predict(X_test)
```

```python
passengerId = test['PassengerId']

columns = ['PassengerId','Survived']
predict_df = pd.DataFrame(
    {'PassengerId': test['PassengerId'],
     'Survived': y_pred},  columns=columns)
predict_df.to_csv('titanic_v4_classifier.csv',index=False)
```
